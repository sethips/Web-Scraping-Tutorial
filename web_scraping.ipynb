```
import requests
website_url = requests.get('https://thegreenestworkforce.ca/index.php/en/schools/').text
```

```
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```
My_table = soup.find('article',{'id':'post-17896'})
My_table
```

```
links = My_table.findAll('a')
links
```

```
colleges = []
for link in links:
    colleges.append(link.get('href'))
    
print(colleges)
```

```
import pandas as pd
df = pd.DataFrame()
df['University'] = colleges
df
```
